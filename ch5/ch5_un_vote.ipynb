```python
from vecutil import *
from matutil import *

f = open('voting_record_dump109.txt')
mylist = list(f)

def create_voting_dict(mylist):
    return { temp_list.split()[0] :  temp_list.split()[3:len(temp_list.split())]  for temp_list in mylist}

voting_dict = create_voting_dict(mylist)

# voting_dict['Myanmar']

voting_matrix = rowdict2mat({country : list2vec([int(value) for value in voting_dict[country]]) for country in voting_dict.keys()})
A = voting_matrix.transpose()



M = A.transpose() * A
print(M)
```
