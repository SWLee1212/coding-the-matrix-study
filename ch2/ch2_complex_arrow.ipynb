```python
import matplotlib.pyplot as plt
import numpy as np


def draw_arrow(a, b):
    ax = plt.axes()
    
    ax.arrow(0,0,a.real,a.imag, head_width=0.05, head_length=0.1, fc='k', ec='k')             
    ax.arrow(a.real, a.imag, b.real, b.imag, head_width=0.05, head_length=0.1, fc='r', ec='r')            
    ax.arrow(0, 0, a.real+b.real, a.imag+b.imag, head_width=0.05, head_length=0.1, fc='b', ec='b')
    
    ax_xmax = max([abs(a.real), abs(b.real), abs(a.real+b.real)])
    ax_ymax = max([abs(a.imag), abs(b.imag), abs(a.imag+b.imag)])
    plt.xlim(-ax_xmax, ax_xmax)
    plt.ylim(-ax_ymax, ax_ymax)
    # recompute the ax.dataLim
    
    ax.relim()
    # update ax.viewLim using the new dataLim
    ax.autoscale_view()
    plt.draw()
    
    plt.grid()
    plt.show()

a = 3+1j
b = 2+2j
draw_arrow(a,b)     
             
a = -1+2j
b = 1-1j
draw_arrow(a,b)
```
